```python
import numpy as np
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

from svgpathtools import svg2paths
```

```python
df_path = pd.read_csv('../data/extracts/2024-06-17/fhnw_campus_kurs2_001/odom.csv')
```

```python
plt.plot(np.linspace(0, 100, len(df_path)), df_path["position_z"].to_list())
plt.title("Height profile of path (recorded)")
plt.grid()
plt.show()
```

```python

svg_path = '../data/2024-06-17/rover_pfad2.svg'
paths, attributes = svg2paths(svg_path)

# Assuming there is only one path in the SVG file
svg_path_data = paths[0]

# Sample the SVG path to get points
num_points = len(df_path)  # Make sure to sample the same number of points as in the CSV
svg_points = []

for i in range(num_points):
    svg_points.append(svg_path_data.point(i / (num_points - 1)))

# Convert to numpy array
svg_points = np.array([(point.real, -point.imag) for point in svg_points])
```

```python
span_x, span_y = svg_points.max(axis=0) - svg_points.min(axis=0)
```

```python
span_xr, span_yr = df_path[["position_x", "position_y"]].max() - df_path[["position_x", "position_y"]].min()

factor_x = 1 / span_x * (6 * 2.13)
factor_y = 1 / span_y * (span_yr)
```

```python

svg_points *= factor_y
```

```python


plt.figure(figsize=(8, 10))
plt.plot(svg_points[:, 0], svg_points[:, 1], label='SVG Path')

plt.plot(df_path["position_x"].iloc[0], df_path["position_y"].iloc[0], 'rx', label="Start",)
plt.plot(df_path["position_x"].iloc[-1], df_path["position_y"].iloc[-1], 'bx', label="End",)

plt.plot(df_path["position_x"].to_list(), df_path["position_y"].to_list(), label="State Estimate")

plt.title("fhnw_campus_kurs2_001")
plt.gca().set_aspect('equal')
plt.legend()
plt.show()
```
